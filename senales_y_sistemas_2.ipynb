## 🔹 2.3 Operaciones de interconexión de sistemas

---

### 1️⃣ ¿Qué significa interconectar sistemas?

<details>
<summary>💡 Ver respuesta</summary>

**Respuesta:**  
Interconectar sistemas significa combinar varios sistemas LTI (Lineales e Invariantes en el Tiempo) para formar uno nuevo.  
Cada conexión afecta cómo se calcula la salida total del conjunto.

**Teoría breve:**  
Las interconexiones más comunes son:
- **Serie:** la salida del primero es la entrada del segundo.  
- **Paralelo:** las salidas se suman.  
- **Realimentación:** una parte de la salida se devuelve a la entrada.  

</details>

---

### 2️⃣ Propiedad distributiva respecto a la suma

<details>
<summary>💡 Ver respuesta</summary>

**Fórmula:**
$$
x[n] * (h_1[n] + h_2[n]) = x[n]*h_1[n] + x[n]*h_2[n]
$$

**Explicación sencilla:**  
Esto significa que si una misma señal pasa por **dos sistemas diferentes en paralelo**, la salida total será la **suma** de las salidas individuales.

**Representación gráfica (paralelo):**

<pre>
       ┌───── h₁[n] ─────┐
x[n] ──┤                 ├──► y[n] = x[n]*h₁[n] + x[n]*h₂[n]
       └───── h₂[n] ─────┘
</pre>

Cada bloque aplica una convolución distinta sobre la misma entrada $x[n]$,  
y la salida total se obtiene sumando los dos resultados.

</details>


### 3️⃣ Propiedad asociativa (sistemas en serie)

<details>
<summary>💡 Ver respuesta</summary>

**Fórmula:**
$$
x[n] * (h_1[n] * h_2[n]) = (x[n]*h_1[n]) * h_2[n]
$$

**Explicación sencilla:**  
En un sistema en **serie**, la salida del primer bloque se convierte en la entrada del segundo.  
La propiedad asociativa indica que el **orden de agrupación no altera el resultado**:  
puedes convolucionar primero $x[n]$ con $h_1[n]$, o $h_1[n]$ con $h_2[n]$, el resultado será igual.

**Representación gráfica (serie):**

<pre>
x[n] ──► [ h₁[n] ] ──► [ h₂[n] ] ──► y[n]
</pre>

Esto equivale a un solo sistema LTI con respuesta al impulso:

$$
h_{eq}[n] = h₁[n] * h₂[n]
$$

**Ejemplo conceptual:**  
Imagina dos filtros: el primero suaviza la señal y el segundo atenúa frecuencias altas.  
Juntos actúan como un solo filtro más complejo, pero la operación total sigue siendo una **convolución**.

</details>





El conjunto puede reemplazarse por un solo sistema equivalente:
$$
h_{eq}[n] = h₁[n] * h₂[n]
$$

**Observación importante:**  
> Si el orden **no** es intercambiable, entonces **al menos uno no es LTI**.

</details>

---

### 4️⃣ Ejemplo gráfico de asociación

<details>
<summary>💡 Ver respuesta</summary>

Si un sistema tiene dos etapas:
- $h_1[n] = (1/2)^n u[n]$
- $h_2[n] = u[n+3]$

Entonces la respuesta equivalente es su **convolución**:
$$
h_{eq}[n] = h_1[n] * h_2[n]
$$

Que se interpreta como un **desplazamiento y ponderación** de $h_1[n]$.

</details>

## ⚙️ 5. Propiedades de los sistemas LTI

---

### 5️⃣ ¿Qué significa que un sistema tenga memoria?

<details>
<summary>💡 Ver respuesta</summary>

**Respuesta:**  
Un sistema **sin memoria** produce una salida que depende solo del valor actual de la entrada.  
En cambio, un sistema **con memoria** depende de valores pasados (o futuros) de la entrada.

**Ejemplo:**  
- Sin memoria: $y[n] = 2x[n]$  
- Con memoria: $y[n] = x[n-1] + x[n]$

**Teoría:**  
Para un sistema LTI sin memoria, la **respuesta al impulso** es:
$$
h[n] = c\,\delta[n]
$$
Porque solo responde al instante actual.

</details>

---

### 6️⃣ ¿Qué es la causalidad en un sistema?

<details>
<summary>💡 Ver respuesta</summary>

**Respuesta:**  
Un sistema **causal** es aquel cuya salida en un instante depende **solo del presente o del pasado** de la entrada, nunca del futuro.

**Condición matemática:**
$$
h[n] = 0, \quad \text{para } n < 0
$$

**Ejemplo:**  
El sistema $y[n]=x[n]+x[n-1]$ es causal.  
Pero $y[n]=x[n+1]$ **no lo es**, porque usa un valor futuro.

</details>

---

### 7️⃣ ¿Qué significa que un sistema sea estable?

<details>
<summary>💡 Ver respuesta</summary>

**Respuesta:**  
Un sistema es **estable** (BIBO: Bounded Input, Bounded Output) si toda entrada limitada produce una salida limitada.

**Condición matemática:**
$$
\sum_{n=-\infty}^{\infty} |h[n]| < \infty
\quad \text{o en continuo} \quad
\int_{-\infty}^{\infty} |h(t)|\,dt < \infty
$$

**Interpretación:**  
El sistema no amplifica indefinidamente la energía de la señal.

**Ejemplo (problema 2.14):**
$$
h(t) = e^{-t}\cos(2t)u(t)
$$
Aquí:
$$
\int_0^{\infty} |e^{-t}\cos(2t)|dt \le \int_0^{\infty} e^{-t}dt = 1
$$
Como la integral es finita, **el sistema es estable**.

</details>

---

### 8️⃣ ¿Qué significa que un sistema sea invertible?

<details>
<summary>💡 Ver respuesta</summary>

**Respuesta:**  
Un sistema LTI es **invertible** si diferentes entradas producen siempre diferentes salidas.  
Eso permite recuperar la entrada a partir de la salida.

**Condición matemática:**
$$
h(t) * h_i(t) = \delta(t)
$$
donde $h_i(t)$ es el **sistema inverso**.

**Ejemplo:**  
Un sistema que multiplica por 2, $y=2x$, tiene como inverso $x=\tfrac{1}{2}y$.

</details>

---

### 9️⃣ ¿Cómo se representa la memoria, causalidad y estabilidad gráficamente?

<details>
<summary>💡 Ver respuesta</summary>

| Propiedad | Descripción | Condición sobre $h(t)$ |
|------------|--------------|------------------------|
| Sin memoria | Depende solo de $x(t)$ actual | $h(t)=c\delta(t)$ |
| Causal | Depende del presente/pasado | $h(t)=0$ para $t<0$ |
| Estable | Responde de forma acotada | $\int |h(t)|dt < ∞$ |
| Invertible | Se puede recuperar $x(t)$ | $h*h_i = \delta(t)$ |

**Intuición:**  
- Si el sistema “recuerda” valores → tiene memoria.  
- Si reacciona antes de recibir señal → no es causal.  
- Si la respuesta crece sin límite → inestable.  
- Si aplasta o pierde información → no invertible.

</details>

---

### 🔟 Ejemplo: Análisis de causalidad y estabilidad

<details>
<summary>💡 Ver respuesta</summary>

**Problema 2.28 (caso discreto):**  
$$
h[n] = (0.8)^n u[n+2]
$$

**Causalidad:**  
Como $u[n+2]$ empieza en $n=-2$, el sistema **no es causal** (usa valores anteriores a $n=0$).

**Estabilidad:**  
$$
\sum_{n=-\infty}^{\infty}|h[n]| = \sum_{n=-2}^{\infty}(0.8)^n
= \frac{(0.8)^{-2}}{1-0.8} = \frac{1.56}{0.2}<\infty
$$  
⇒ **El sistema es estable.**

</details>

### 4️⃣ Interconexión en realimentación (feedback)

<details>
<summary>💡 Ver respuesta</summary>

**Idea principal:**  
En una **realimentación**, parte de la salida se devuelve a la entrada del sistema.  
Esto modifica el comportamiento del sistema total, pudiendo **amplificar, estabilizar o filtrar** la señal.

---

**Diagrama conceptual (realimentación unitaria):**

<pre>
         ┌─────────────────────┐
         │                     ▼
x[n] ──►(+)──►[ h₁[n] ]──► y[n]
          ▲
          │
        [ h₂[n] ]
</pre>

---

**Relación matemática:**

Si la salida $y[n]$ se retroalimenta a la entrada a través de un sistema $h_2[n]$,  
y la señal directa pasa por $h_1[n]$, la ecuación general es:

$$
y[n] = x[n]*h_1[n] + y[n]*h_2[n]*h_1[n]
$$

Reordenando:

$$
y[n] (1 - h_1[n]*h_2[n]) = x[n]*h_1[n]
$$

Por tanto, el **sistema equivalente** (sin lazo) tiene respuesta al impulso:

$$
h_{eq}[n] = \frac{h_1[n]}{1 - h_1[n]*h_2[n]}
$$

---

**Explicación intuitiva:**  
- Si $h_2[n]$ es positiva, la realimentación **aumenta** el efecto del sistema.  
- Si $h_2[n]$ es negativa, la realimentación **reduce** o **estabiliza** la salida.  
- En muchos sistemas físicos (como control automático o amplificadores), la realimentación negativa ayuda a **evitar saturaciones** o **aumentar precisión**.

---

**Ejemplo numérico:**  
Supón que $h_1[n] = a^n u[n]$ y $h_2[n] = k\,\delta[n]$.  
Entonces:

$$
h_{eq}[n] = \frac{a^n u[n]}{1 - k a^n u[n]}
$$

Si $|k a| < 1$, el sistema **es estable**.

</details>


## ⚙️ 5️⃣ Propiedades de los Sistemas LTI

Los sistemas **LTI** (Lineales e Invariantes en el Tiempo) se caracterizan por cuatro propiedades fundamentales:  
👉 **Memoria**, **Causalidad**, **Estabilidad (BIBO)** e **Invertibilidad**.  
Estas propiedades determinan cómo el sistema responde ante cualquier señal de entrada.

---

### 1️⃣ ¿Qué significa que un sistema tenga memoria?

<details>
<summary>💡 Ver respuesta</summary>

**Definición:**  
Un sistema tiene **memoria** si su salida en un instante depende de valores pasados o futuros de la entrada.

**En cambio**, un sistema es **sin memoria** si $y(t)$ depende únicamente del valor actual de $x(t)$.

**Ejemplos:**
- Sin memoria → $y(t)=3x(t)$  
- Con memoria → $y(t)=x(t-2)$ o $y(t)=x(t-1)+x(t)$

**Interpretación intuitiva:**  
Un sistema con memoria “recuerda” valores previos de la entrada.  
Un sistema sin memoria responde **instantáneamente**.

**Condición sobre la respuesta al impulso:**
$$
h(t) = c\delta(t) \Rightarrow \text{Sistema sin memoria}
$$

</details>

---

### 2️⃣ ¿Qué significa causalidad?

<details>
<summary>💡 Ver respuesta</summary>

**Definición:**  
Un sistema es **causal** si la salida en un instante solo depende del presente y del pasado de la entrada, **nunca del futuro**.

**Condición matemática:**
$$
h(t) = 0 \quad \text{para } t < 0
$$

**Ejemplos:**
- Causal → $y(t) = x(t-1)$ o $y(t)=x(t)$  
- No causal → $y(t)=x(t+2)$ (usa valores futuros)

**Intuición:**  
Un sistema físico real **siempre es causal**, porque no puede reaccionar antes de recibir la señal.

**Visualización:**

<pre>
Tiempo:  ← pasado ─── 0 ─── futuro →
h(t):    |••••••••|0|0|0|→     (causal)
</pre>

</details>

---

### 3️⃣ ¿Qué es la estabilidad BIBO?

<details>
<summary>💡 Ver respuesta</summary>

**Definición:**  
BIBO = *Bounded Input → Bounded Output*.  
Un sistema es **estable** si toda entrada acotada produce una salida acotada.

**Condición matemática (continua):**
$$
\int_{-\infty}^{\infty} |h(t)|\,dt < \infty
$$

**Condición (discreta):**
$$
\sum_{n=-\infty}^{\infty} |h[n]| < \infty
$$

**Ejemplo:**  
$h(t)=e^{-t}u(t)$  
$$
\int_0^\infty e^{-t}\,dt = 1 < \infty \Rightarrow \text{estable}
$$

**Contraejemplo:**  
$h(t)=e^{t}u(t)$  
$$
\int_0^\infty e^{t}\,dt = \infty \Rightarrow \text{inestable}
$$

**Intuición:**  
Si $h(t)$ tiene una cola infinita que no decae, el sistema acumula energía y “explota”.

</details>

---

### 4️⃣ ¿Qué significa que un sistema sea invertible?

<details>
<summary>💡 Ver respuesta</summary>

**Definición:**  
Un sistema es **invertible** si su salida determina de forma única su entrada, es decir, no pierde información.

**Condición matemática:**
$$
h(t) * h_i(t) = \delta(t)
$$
donde $h_i(t)$ es la **respuesta impulsional del sistema inverso**.

**Ejemplo:**  
Sistema: $y(t)=2x(t)$  
Inverso: $x(t)=\tfrac{1}{2}y(t)$  
Entonces:
$$
h(t)=2\delta(t), \quad h_i(t)=\frac{1}{2}\delta(t)
$$
y
$$
h*h_i = \delta(t)
$$

**Interpretación:**  
Si el sistema altera la amplitud o retrasa la señal, aún puede invertirse.  
Pero si la recorta, satura o pierde información, no.

</details>

---

### 5️⃣ ¿Cómo se combinan estas propiedades?

<details>
<summary>💡 Ver respuesta</summary>

| Propiedad | Condición sobre $h(t)$ | Ejemplo | Interpretación |
|:----------|:------------------------|:--------|:----------------|
| **Memoria** | $h(t) \neq c\delta(t)$ | $y(t)=x(t-1)$ | Usa valores pasados |
| **Causalidad** | $h(t)=0$ para $t<0$ | $h(t)=e^{-t}u(t)$ | No anticipa la entrada |
| **Estabilidad** | $\int |h(t)|dt < \infty$ | $h(t)=e^{-t}u(t)$ | Salida acotada |
| **Invertibilidad** | $h*h_i = \delta(t)$ | $h=2\delta(t)$ | Recupera la entrada |

**Resumen mental:**  
- Memoria → “¿Recuerda el pasado?”  
- Causalidad → “¿Actúa antes de tiempo?”  
- Estabilidad → “¿Explota o se mantiene controlado?”  
- Invertibilidad → “¿Pierde información?”

</details>

### 6️⃣ Mini test tipo examen

<details>
<summary>🧠 Preguntas rápidas</summary>

1. **¿Cuándo es estable un sistema LTI?**  
   Cuando $\int |h(t)|dt < \infty$.  

2. **¿Puede ser un sistema sin memoria e inestable?**  
   Sí, si multiplica por un número muy grande ($y=10^6x$).  

3. **¿Qué tipo de sistema tiene $h(t)=2\delta(t)$?**  
   Sin memoria, causal, estable e invertible.

</details>

### 7️⃣ Ejemplo práctico: ¿Causal? ¿Estable?

<details>
<summary>💡 Ver respuesta</summary>

**Ejemplo 2.14 (del libro):**  
$h(t)=e^{-t}\cos(2t)u(t)$

1️⃣ **Causalidad:**  
Como $u(t)$ es 0 para $t<0$, → **Causal**.  

2️⃣ **Estabilidad:**  
$$
\int_{-\infty}^{\infty}|h(t)|dt = \int_0^\infty |e^{-t}\cos(2t)|dt \le \int_0^\infty e^{-t}dt = 1
$$  
La integral es finita → **Estable**.

**Conclusión:**  
Sistema **LTI causal y estable**.  
Si la integral fuera infinita, significaría que el sistema “acumula” respuesta sin límite.

</details>

# ⚙️ Tema 6 — Sistemas LTI definidos por Ecuaciones Diferenciales Lineales (EDL) o en Diferencias

---

## 🧩 1️⃣ ¿Qué es una Ecuación Diferencial Lineal (EDL)?

<details>
<summary>💡 Ver respuesta</summary>

**Definición:**  
Una **EDL** describe la relación entre una **entrada $x(t)$**, su **salida $y(t)$**, y sus **derivadas**.  
Es una forma matemática de representar un sistema **continuo y LTI**.

**Forma general:**
$$
a_N \frac{d^N y(t)}{dt^N} + a_{N-1}\frac{d^{N-1} y(t)}{dt^{N-1}} + \dots + a_0 y(t)
= b_M \frac{d^M x(t)}{dt^M} + \dots + b_0 x(t)
$$

**Interpretación:**  
Cada término de la ecuación muestra cómo el sistema combina **versiones derivadas** de la entrada y salida.  
Los **coeficientes $a_k, b_k$ son constantes**, por eso se llama *de coeficientes constantes*.

**Ejemplo típico:**  
$$
\frac{dy(t)}{dt} + 3y(t) = 2x(t)
$$
→ Sistema de primer orden (porque hay una derivada de $y(t)$).

</details>

---

## 🧮 2️⃣ ¿Qué representa físicamente una EDL?

<details>
<summary>💡 Ver respuesta</summary>

**Respuesta:**  
Representa el **comportamiento dinámico** de un sistema:  
cómo la salida responde a cambios en la entrada, considerando **almacenamiento y disipación de energía**.

**Ejemplo real:**
- En un circuito RC: $\displaystyle \frac{dy}{dt} + \frac{1}{RC}y = \frac{1}{RC}x$  
- En un sistema masa-resorte: $m\ddot{y}+c\dot{y}+ky = F(t)$  

**Intuición:**  
Los coeficientes ($a_k$, $b_k$) determinan cómo el sistema “recuerda” el pasado o “resiste” cambios rápidos.

</details>

---

## 🔹 3️⃣ ¿Qué es una ecuación en diferencias lineal?

<details>
<summary>💡 Ver respuesta</summary>

**Definición:**  
Una **ecuación en diferencias lineal** es la versión *discreta* de una EDL.  
Relaciona la salida $y[n]$ con sus **valores pasados** y con los valores pasados de la entrada $x[n]$.

**Forma general:**
$$
y[n] + a_1y[n-1] + \dots + a_Ny[n-N] = b_0x[n] + b_1x[n-1] + \dots + b_Mx[n-M]
$$

**Ejemplo:**
$$
y[n] = 0.5y[n-1] + x[n]
$$

**Interpretación:**  
Cada nuevo valor de salida se calcula con una **combinación lineal** de salidas pasadas y entradas actuales o anteriores.  
Es un sistema **recursivo**, común en filtros digitales.

</details>

---

## 📈 4️⃣ ¿Cómo se obtiene la respuesta al impulso $h(t)$ o $h[n]$?

<details>
<summary>💡 Ver respuesta</summary>

**Idea:**  
La **respuesta al impulso** se obtiene aplicando una **entrada $\delta$** al sistema y resolviendo la ecuación.

**Ejemplo continuo:**
$$
\frac{dy(t)}{dt} + 3y(t) = \delta(t)
$$

**Solución:**  
$$
h(t) = e^{-3t}u(t)
$$
→ sistema **causal y estable**.

**Ejemplo discreto:**
$$
y[n] = 0.5y[n-1] + \delta[n]
$$
**Solución:**  
$$
h[n] = (0.5)^n u[n]
$$

**Interpretación:**  
Cada respuesta al impulso describe cómo el sistema reacciona a un estímulo instantáneo.

</details>

---

## 🧠 5️⃣ ¿Por qué estos sistemas son LTI?

<details>
<summary>💡 Ver respuesta</summary>

Porque cumplen las dos condiciones:

1. **Linealidad:**  
   Si $x_1, x_2$ generan $y_1, y_2$,  
   entonces $a_1x_1 + a_2x_2 \Rightarrow a_1y_1 + a_2y_2$.

2. **Invarianza temporal:**  
   Si se retrasa la entrada $x(t-t_0)$,  
   la salida también se retrasa: $y(t-t_0)$.

**Ejemplo:**  
El sistema $\frac{dy}{dt}+2y=x$ cumple ambas:  
no depende explícitamente del tiempo y combina entradas de forma lineal.

</details>

---

## 🧩 6️⃣ ¿Cómo se comprueba la estabilidad en una EDL o ecuación en diferencias?

<details>
<summary>💡 Ver respuesta</summary>

**Condición en continuo:**
$$
\text{El sistema es estable si todas las raíces del polinomio } a(s)
\text{ tienen parte real negativa.}
$$

**Condición en discreto:**
$$
\text{El sistema es estable si todas las raíces del polinomio } a(z)
\text{ están dentro del círculo unitario.}
$$

**Ejemplo:**
$$
y[n] = 0.8y[n-1] + x[n]
\Rightarrow a(z)=1-0.8z^{-1}
$$
Raíz en $z=0.8 < 1$ → **estable**.

</details>

## 📘 7️⃣ Ejemplo completo (Problema 2.18)

<details>
<summary>🧩 Ver desarrollo paso a paso</summary>

**Enunciado:**
$$
y[n] - 0.5y[n-1] = x[n]
$$

1️⃣ **Identificar:**  
Ecuación en diferencias lineal con coeficiente $a_1=0.5$.

2️⃣ **Respuesta al impulso:**  
Entrada $x[n]=\delta[n]$.  
Aplicamos la ecuación:
$$
h[n] - 0.5h[n-1] = \delta[n]
$$

3️⃣ **Desarrollo:**
- Para $n=0$: $h[0]-0.5h[-1]=1 \Rightarrow h[0]=1$
- Para $n=1$: $h[1]-0.5h[0]=0 \Rightarrow h[1]=0.5$
- Para $n=2$: $h[2]-0.5h[1]=0 \Rightarrow h[2]=0.25$

**Conclusión:**
$$
h[n]=(0.5)^n u[n]
$$
→ Sistema **causal y estable** (la secuencia decae).

</details>

## 📒 🔟 Ejemplo (Problema 2.26)

<details>
<summary>⚙️ Interconexión y equivalencia</summary>

**Sistema 1:** $h_1(t)=e^{-t}u(t)$  
**Sistema 2:** $h_2(t)=u(t-1)$  

Conexión en **serie** →  
$$

$$

$$

**Interpretación:**  
- La salida empieza en $t=1$ → causal.  
- Decae exponencialmente → estable.  
- Representa un **retardo + filtro exponencial**.

</details>

# 🧭 Resumen General — Tema 2: Sistemas LTI

---

## 🎛️ 1️⃣ Estructura general de un sistema LTI

<pre>
        Entrada x(t) ──► [ h(t) ] ──► Salida y(t)
</pre>

**Relación fundamental:**
$$
y(t) = x(t) * h(t) = \int_{-\infty}^{\infty} x(\tau)\,h(t-\tau)\,d\tau
$$

- $x(t)$ → Señal de entrada  
- $h(t)$ → Respuesta al impulso  
- $y(t)$ → Señal de salida  

**Interpretación:**  
El sistema LTI genera la salida como la **suma ponderada y desplazada** de copias de $h(t)$ según los valores de $x(t)$.

---

## 🔄 2️⃣ Operaciones de interconexión

| Tipo | Descripción | Fórmula | Interpretación |
|------|--------------|----------|----------------|
| **Serie** | Salida de uno entra al siguiente | $h_{eq}=h_1*h_2$ | Orden no importa (asociativa) |
| **Paralelo** | Misma entrada, se suman salidas | $x*(h_1+h_2)$ | Suma distributiva |
| **Realimentación** | Parte de la salida vuelve a la entrada | $h_{eq}=\frac{h_1}{1-h_1h_2}$ | Aumenta o estabiliza la salida |

**Diagramas:**

<pre>
Serie:          x → [h₁] → [h₂] → y
Paralelo:       ┌→[h₁]─┐
           x →──┤      ├──→ y
                └→[h₂]─┘
Realimentación: x →(+)→[h₁]→ y
                  ↑
                 [h₂]
</pre>

---

## ⚙️ 3️⃣ Propiedades básicas de los sistemas LTI

| Propiedad | Condición sobre $h(t)$ | Ejemplo | Intuición |
|------------|------------------------|----------|------------|
| **Memoria** | $h(t)\neq c\delta(t)$ | $h(t)=u(t-1)$ | Depende del pasado |
| **Causalidad** | $h(t)=0$ para $t<0$ | $h(t)=e^{-t}u(t)$ | No anticipa la entrada |
| **Estabilidad (BIBO)** | $\int |h(t)|dt<\infty$ | $h(t)=e^{-t}u(t)$ | No explota |
| **Invertibilidad** | $h*h_i=\delta(t)$ | $h(t)=2\delta(t)$ | Se puede deshacer |
| **Linealidad** | $T(a_1x_1+a_2x_2)=a_1T(x_1)+a_2T(x_2)$ | – | Responde proporcionalmente |
| **Invarianza temporal** | Si $x(t-t_0)\Rightarrow y(t-t_0)$ | – | Respuesta no cambia en el tiempo |

**Esquema visual:**

<pre>
           ┌──────────────┬─────────────────────────────┐
           │ Propiedad    │  Cómo se ve en h(t)         │
           ├──────────────┼─────────────────────────────┤
           │ Sin memoria  │  h(t)=cδ(t)                 │
           │ Causal       │  h(t)=0 para t<0            │
           │ Estable      │  ∫|h(t)|dt < ∞              │
           │ Invertible   │  h*h_i=δ(t)                 │
           └──────────────┴─────────────────────────────┘
</pre>

---

## 📘 4️⃣ Sistemas definidos por ecuaciones diferenciales o en diferencias

**Forma continua (EDL):**
$$
a_N y^{(N)}(t)+\dots+a_1y'(t)+a_0y(t)
= b_Mx^{(M)}(t)+\dots+b_0x(t)
$$

**Forma discreta (EDL en diferencias):**
$$
y[n]+a_1y[n-1]+\dots+a_Ny[n-N]
= b_0x[n]+b_1x[n-1]+\dots+b_Mx[n-M]
$$

**Ejemplos:**
| Tipo | Ecuación | Respuesta al impulso |
|------|-----------|----------------------|
| Continua | $\frac{dy}{dt}+3y=x$ | $h(t)=e^{-3t}u(t)$ |
| Discreta | $y[n]=0.8y[n-1]+x[n]$ | $h[n]=(0.8)^n u[n]$ |
| No recursiva | $y[n]=x[n]+2x[n-1]$ | $h[n]=\delta[n]+2\delta[n-1]$ |

**Propiedades:**
- **Causales** si sólo usan valores presentes/pasados.  
- **Estables** si las raíces tienen parte real < 0 (en continuo) o $|z|<1$ (en discreto).  
- **Con memoria**, salvo que todos los términos sean instantáneos.

---

## 🔍 6️⃣ Estabilidad: interpretación con la integral

**Condición:**  
$$
\int_{-\infty}^{\infty}|h(t)|\,dt < \infty
$$

**Significado:**  
La respuesta impulsional debe tener **energía total finita**.  
Si la integral diverge, el sistema amplifica indefinidamente la señal.

**Ejemplo práctico (2.14):**
$$
h(t)=e^{-t}\cos(2t)u(t) \Rightarrow \int_0^\infty e^{-t}dt=1
$$
✅ → Estable.

# 🎵 Tema 3 – Series de Fourier de Señales Periódicas

---

## 🧭 1️⃣ ¿Por qué aparece la Serie de Fourier en Señales y Sistemas?

<details>
<summary>💡 Ver explicación</summary>

Hasta ahora hemos estudiado **sistemas LTI** y cómo responden a cualquier señal.  
Pero muchas señales reales son **periódicas** (repetitivas): como el sonido, la corriente alterna, o un pulso digital.  

Jean-Baptiste Fourier descubrió que **cualquier señal periódica puede representarse como suma de senos y cosenos**.

👉 Esto se llama **Serie de Fourier**.  

Matemáticamente:
$$
x(t) = \sum_{k=-\infty}^{\infty} a_k e^{jk\omega_0 t}
$$

**Intuición:**  
En lugar de estudiar una señal complicada, la descomponemos en sus “ingredientes” más simples: **ondas puras (frecuencias)**.

Cada $a_k$ indica **cuánto aporta** cada frecuencia $k\omega_0$ a la forma total de la señal.

</details>

---

## 🎯 2️⃣ ¿Qué significa “descomponer una señal”?

<details>
<summary>💡 Ver respuesta</summary>

**Descomponer** una señal significa expresarla como **suma de componentes más simples** (senos, cosenos o exponenciales).  

Ejemplo intuitivo:
> Una melodía compleja puede dividirse en notas puras (frecuencias).

Del mismo modo, una señal $x(t)$ se puede “traducir” al idioma de la frecuencia mediante sus **coeficientes de Fourier $a_k$**.

---

**Fórmula general (forma exponencial):**
$$
x(t) = \sum_{k=-\infty}^{\infty} a_k e^{jk\omega_0 t}
$$

**Y sus coeficientes:**
$$
a_k = \frac{1}{T_0}\int_{T_0} x(t)e^{-jk\omega_0 t}\,dt
$$

Donde:  
- $T_0$ → período fundamental  
- $\omega_0 = \dfrac{2\pi}{T_0}$ → frecuencia angular fundamental  

</details>

---

## 🎶 3️⃣ ¿Para qué sirve la Serie de Fourier?

<details>
<summary>💡 Ver explicación</summary>

Sirve para **entender y analizar señales periódicas** de una forma más práctica:

| Aplicación | Qué permite |
|-------------|-------------|
| 🔊 Audio / Sonido | Descomponer en tonos puros (frecuencias audibles) |
| ⚡ Electricidad | Analizar armónicos en corriente alterna |
| 🧮 Matemática | Simplificar integrales y ecuaciones diferenciales |
| 🖥️ Ingeniería digital | Sintetizar o comprimir señales (Fourier → transformadas) |
| 🎚️ Filtros LTI | Saber cómo un sistema modifica cada frecuencia |

**Ejemplo intuitivo:**  
Un altavoz no responde igual a un tono grave que a uno agudo.  
Con Fourier, puedes estudiar cómo responde el sistema a **cada frecuencia por separado**.

</details>

---

## 📊 4️⃣ ¿Qué representa cada coeficiente $a_k$?

<details>
<summary>💡 Ver respuesta</summary>

Cada $a_k$ es un **número complejo** que contiene:
- La **amplitud** de la frecuencia $k\omega_0$  
- Su **fase** (desplazamiento en el tiempo)

**Fórmula:**
$$
a_k = |a_k|e^{j\angle a_k}
$$

**Interpretación visual:**
- $|a_k|$ → altura del pico en el espectro de amplitud  
- $\angle a_k$ → desplazamiento horizontal (fase)  

📈 Si graficas todos los $|a_k|$ en función de $k$, obtienes el **espectro de amplitud** de la señal.

</details>

---

## 🧩 5️⃣ ¿Por qué usamos exponenciales complejas y no solo senos y cosenos?

<details>
<summary>💡 Ver explicación</summary>

Porque las exponenciales complejas **simplifican mucho las matemáticas**.  

Gracias a la identidad de Euler:
$$
e^{j\theta} = \cos\theta + j\sin\theta
$$

Así, una sola función $e^{j\omega t}$ incluye **dos componentes reales** (coseno y seno).  
Esto permite representar señales de manera más compacta y manipularlas fácilmente con sistemas LTI.

**Además:**
- Las exponenciales son **autofunciones** de los sistemas LTI:  
  si entras con $e^{j\omega t}$, la salida es proporcional a la misma exponencial.  
  $$T\{e^{j\omega t}\}=H(j\omega)e^{j\omega t}$$  
  donde $H(j\omega)$ es la **respuesta en frecuencia** del sistema.

</details>

---

## ⚡ 6️⃣ Relación con sistemas LTI

<details>
<summary>💡 Ver respuesta</summary>

Un sistema LTI responde a cada componente $e^{j\omega t}$ **de forma independiente**.  
Por eso, si descompones $x(t)$ en Fourier, puedes obtener fácilmente $y(t)$ aplicando la ganancia de cada frecuencia.

**Matemáticamente:**
$$
y(t)=\sum_{k=-\infty}^{\infty} H(jk\omega_0)a_k e^{jk\omega_0 t}
$$

**Interpretación:**  
El sistema **multiplica cada frecuencia** por su factor $H(jk\omega_0)$ (su ganancia y fase).  
Por eso, la Serie de Fourier es **la herramienta perfecta para estudiar sistemas LTI periódicos**.

</details>

---

## 🧠 7️⃣ Dudas más comunes entre los estudiantes

<details>
<summary>💬 Ver lista</summary>

1. **“¿Por qué hay infinitas exponenciales?”**  
   → Porque se necesita una base infinita para reconstruir cualquier forma periódica.

2. **“¿Por qué los $a_k$ son complejos si la señal es real?”**  
   → Porque se usan exponenciales complejas. En señales reales, $a_{-k}=a_k^*$ (simetría hermítica).

3. **“¿Qué diferencia hay entre Serie y Transformada de Fourier?”**  
   → La serie se aplica a **señales periódicas**; la transformada, a **señales no periódicas**.

4. **“¿Qué significa $\omega_0$?”**  
   → Es la frecuencia angular básica: cada armónico es múltiplo de $\omega_0$.

5. **“¿Por qué se usa $1/T_0$ en la fórmula de $a_k$?”**  
   → Para **normalizar** la energía del período y que la reconstrucción sea exacta.

6. **“¿Qué pasa si la señal no cumple las condiciones de Dirichlet?”**  
   → La serie converge, pero puede presentar “saltos” o **efecto Gibbs** (oscilaciones en los bordes).

7. **“¿Cómo saber si una señal tiene Serie de Fourier?”**  
   → Si cumple las **condiciones de Dirichlet**:  
      - $x(t)$ tiene número finito de discontinuidades  
      - Finito número de máximos/mínimos  
      - Energía finita en un período

</details>

<details>
<summary>🧩 Haz clic para practicar</summary>

   → La contribución de la frecuencia $k\omega_0$ a la señal total.

# 🧩 Explicación detallada de cada término en la Serie de Fourier

---

## 🌊 1️⃣ La fórmula general

$$
x(t) = \sum_{k=-\infty}^{\infty} a_k e^{j k \omega_0 t}
$$

Esta expresión dice que **una señal periódica $x(t)$** se puede reconstruir como la **suma infinita** de pequeñas ondas (exponenciales) que vibran a distintas frecuencias.

---

## 🔹 2️⃣ ¿Qué representa cada símbolo?

| Símbolo | Se llama | Qué significa en palabras simples |
|:--|:--|:--|
| $x(t)$ | Señal original | Es la señal periódica que queremos representar. |
| $\sum_{k=-\infty}^{\infty}$ | Suma infinita | Sumamos todas las frecuencias (armónicos) positivas y negativas. |
| $a_k$ | Coeficiente complejo | Indica cuánto contribuye la frecuencia $k\omega_0$ en la señal (su peso). |
| $e^{j k \omega_0 t}$ | Exponencial compleja | Es la onda pura que vibra a la frecuencia $k\omega_0$. |
| $\omega_0$ | Frecuencia angular fundamental | Es la frecuencia más baja de la señal (el “ritmo base”). |
| $k$ | Armónico | Es un número entero (0, ±1, ±2, …) que indica múltiplos de $\omega_0$. |

---

## 🔸 3️⃣ El período y la frecuencia angular

**Período fundamental ($T_0$):**
> Es el tiempo que tarda la señal en repetirse.

$$
T_0 = \text{período fundamental}
$$

**Frecuencia angular fundamental ($\omega_0$):**
> Es la velocidad con la que gira cada armónico en radianes/segundo.

$$
\omega_0 = \frac{2\pi}{T_0}
$$

**Ejemplo:**
Si una señal se repite cada 1 s ($T_0=1$),  
entonces $\omega_0 = 2\pi$ rad/s.  
El armónico 2 vibrará el doble de rápido: $2\omega_0 = 4\pi$ rad/s.

---

## ⚙️ 4️⃣ El término $a_k$ (coeficiente de Fourier)

### 🧮 Fórmula:

$$
a_k = \frac{1}{T_0}\int_{T_0} x(t)e^{-j k \omega_0 t}\,dt
$$

### 💬 Explicación paso a paso:
1. Multiplicamos $x(t)$ por $e^{-j k \omega_0 t}$ → esto “aísla” la frecuencia $k$.
2. Integramos sobre un período → calculamos su contribución total.
3. Dividimos por $T_0$ → hacemos el promedio.

**Resultado:**  
$a_k$ te dice **cuánto de la frecuencia $k\omega_0$** hay dentro de la señal $x(t)$.

### 🧠 Intuición:
- Si $a_k$ es grande → esa frecuencia tiene mucho peso.  
- Si $a_k = 0$ → esa frecuencia no está presente.

---

## 🎚️ 5️⃣ El término $e^{j k \omega_0 t}$ (armónico)

Este término representa una **onda base** o **armónico**.  
Cada armónico tiene su propia frecuencia $k\omega_0$.

- Si $k=0$ → $e^{j·0·t}=1$ → es la **componente continua (DC)** (el promedio).  
- Si $k=1$ → frecuencia fundamental → tono base.  
- Si $k=2,3,4...$ → armónicos superiores (duplican, triplican la frecuencia).  

**Ejemplo visual (imaginario):**

<pre>
x(t) = (onda lenta) + (onda media) + (onda rápida)
</pre>

Cada una de esas ondas es un término $e^{j k \omega_0 t}$.

---

## 💡 6️⃣ El índice $k$ (armónico positivo y negativo)

| Valor de $k$ | Qué representa |
|:--|:--|
| $k=0$ | Componente DC (promedio de la señal) |
| $k>0$ | Frecuencias positivas (rotan hacia un lado en el plano complejo) |
| $k<0$ | Frecuencias negativas (rotan al lado contrario) |

Para **señales reales**, los armónicos positivos y negativos se compensan:
$$
a_{-k} = a_k^*
$$
(de modo que la señal final es real).

---

## 🔸 7️⃣ ¿Por qué aparecen números complejos?

Porque las exponenciales complejas ($e^{j\omega t}$) permiten representar **al mismo tiempo** un seno y un coseno:

$$
e^{j\omega t} = \cos(\omega t) + j\sin(\omega t)
$$

Esto simplifica los cálculos:
- Las sumas y derivadas se vuelven más fáciles.
- En sistemas LTI, las exponenciales **no cambian de forma** → son *autofunciones*.

---

## 🧮 8️⃣ Reconstrucción de la señal

Una vez calculados todos los $a_k$, podemos **reconstruir la señal** sumando todos los armónicos:

$$
x(t) = \sum_{k=-\infty}^{\infty} a_k e^{j k \omega_0 t}
$$

Si $x(t)$ es real, también puede escribirse con senos y cosenos:

$$
x(t) = a_0 + \sum_{k=1}^{\infty} [A_k\cos(k\omega_0 t) + B_k\sin(k\omega_0 t)]
$$

donde:
- $A_k = 2\Re\{a_k\}$  
- $B_k = -2\Im\{a_k\}$

---

## 📊 9️⃣ Interpretación física

Cada $a_k$ es como el **volumen (amplitud)** de una nota musical.  
Sumando todas las notas (frecuencias) recuperas la melodía original.

**Ejemplo real:**
Una señal cuadrada puede representarse como:
$$
x(t) = \frac{4}{\pi}[\sin(\omega_0 t) + \frac{1}{3}\sin(3\omega_0 t) + \frac{1}{5}\sin(5\omega_0 t)+\dots]
$$

Es decir, una onda cuadrada = suma infinita de senos impares.

---

## 🧠 10️⃣ Conclusión intuitiva

| Elemento | Pregunta clave | Respuesta |
|:--|:--|:--|
| $x(t)$ | ¿Qué señal estoy analizando? | La original, periódica |
| $a_k$ | ¿Cuánto aporta cada frecuencia? | Peso de cada armónico |
| $e^{j k \omega_0 t}$ | ¿Qué forma tiene cada armónico? | Onda compleja pura |
| $\omega_0$ | ¿Cuál es la frecuencia base? | $2\pi/T_0$ |
| $\sum$ | ¿Qué hago con todo eso? | Sumar todas las ondas |

En resumen:
> La Serie de Fourier es una **receta** que mezcla infinitas ondas simples para cocinar cualquier señal periódica.

# 🧩 Problema 3.3 — Serie de Fourier en forma exponencial

> **Enunciado**  
> Sea  
> $$x(t) = 1 + 2\cos(4t) + 3\sin(6t).$$  
> Obtener su **Serie de Fourier en forma exponencial**  
> $$x(t) = \sum_{k=-\infty}^{\infty} a_k\, e^{j k \omega_0 t},$$  
> determinando la frecuencia fundamental $\omega_0$, el período $T_0$ y los coeficientes $a_k$.

---

## 🧠 Teoría previa

Toda señal periódica puede expresarse como combinación de **exponenciales complejas** $e^{j k \omega_0 t}$.

Cada coeficiente $a_k$ mide **cuánto contribuye la frecuencia** $k\omega_0$ a la señal.

Definiciones fundamentales:

$$
a_k = \frac{1}{T_0} \int_{T_0} x(t) e^{-j k \omega_0 t} dt,
\qquad
\omega_0 = \frac{2\pi}{T_0}.
$$

---

## 🧮 Paso 1 — Identificar frecuencias presentes

La señal contiene las frecuencias angulares:

$$
\omega_1 = 4, \qquad \omega_2 = 6.
$$

Como son **conmensurables**, existe una **frecuencia fundamental**:

$$
\omega_0 = \gcd(4,6) = 2, \qquad T_0 = \frac{2\pi}{\omega_0} = \pi.
$$

> Si las frecuencias no fueran conmensurables, la señal sería **aperiódica**  
> y **no tendría Serie de Fourier**.

---

## 🧮 Paso 2 — Reescribir usando exponenciales de $\omega_0$

Recordemos las identidades de Euler:

$$
\cos(4t) = \tfrac{1}{2}\big(e^{j4t} + e^{-j4t}\big),
\qquad
\sin(6t) = \tfrac{1}{2j}\big(e^{j6t} - e^{-j6t}\big).
$$

Sustituyendo en $x(t)$:

$$
\begin{aligned}
x(t)
&= 1 + 2\cos(4t) + 3\sin(6t) \\[2mm]
&= 1 + 2\left[\tfrac{1}{2}(e^{j4t}+e^{-j4t})\right]
   + 3\left[\tfrac{1}{2j}(e^{j6t}-e^{-j6t})\right] \\[2mm]
&= 1 + e^{j4t} + e^{-j4t}
   + \frac{3}{2j}e^{j6t} - \frac{3}{2j}e^{-j6t}.
\end{aligned}
$$

Como $\omega_0 = 2$, escribimos $4 = 2\cdot2$ y $6 = 3\cdot2$:

$$
x(t) = 1 + e^{j(2\cdot2)t} + e^{-j(2\cdot2)t}
       + \frac{3}{2j}e^{j(3\cdot2)t} - \frac{3}{2j}e^{-j(3\cdot2)t}.
$$

---

## 🧩 Paso 3 — Identificar los coeficientes $a_k$

Comparando con la forma general $x(t) = \sum a_k e^{j k \omega_0 t}$  
(con $\omega_0 = 2$):

| $k$ | $a_k$ |
|:--:|:--:|
| $0$ | $1$ |
| $+2$ | $1$ |
| $-2$ | $1$ |
| $+3$ | $\displaystyle \frac{3}{2j} = -j\,\frac{3}{2}$ |
| $-3$ | $\displaystyle -\frac{3}{2j} = +j\,\frac{3}{2}$ |
| resto | $0$ |

---

## 🧮 Paso 4 — Verificar que la señal es real

Para señales reales, debe cumplirse:

$$
a_{-k} = a_k^*.
$$

Y efectivamente:

$$
a_{-2}=a_2=1, \qquad a_{-3}=a_3^*.
$$

---

## 🧾 Paso 5 — Reconstrucción de la señal

$$
\begin{aligned}
x(t)
&= a_0 + a_2 e^{j2\omega_0 t} + a_{-2} e^{-j2\omega_0 t}
     + a_3 e^{j3\omega_0 t} + a_{-3} e^{-j3\omega_0 t} \\[2mm]
&= 1 + 2\cos(4t) + 3\sin(6t).
\end{aligned}
$$

---

## ✅ Resultado final (problema 3.3)

$$
\boxed{
\omega_0 = 2, \quad
T_0 = \pi, \quad
a_0 = 1, \;
a_{\pm 2} = 1, \;
a_{+3} = -j\frac{3}{2}, \;
a_{-3} = j\frac{3}{2}.
}
$$

---

## 💡 Comentario teórico

- Este problema demuestra la **importancia de encontrar una frecuencia fundamental común** antes de calcular la Serie de Fourier.  
- Si las frecuencias no comparten un múltiplo común → la señal no es periódica.  
- La **forma exponencial** facilita la lectura directa de amplitudes y fases de los armónicos.

---

### 🧩 Mini–test tipo examen

1️⃣ ¿Qué representa cada $a_k$? → La amplitud y fase de la frecuencia $k\omega_0$.  
2️⃣ ¿Qué pasa si las frecuencias son inconmensurables? → La señal no es periódica.  
3️⃣ ¿Cuál es $a_0$? → Es la **componente DC** (valor medio).  
4️⃣ ¿Qué relación hay entre $a_{+k}$ y $a_{-k}$ en señales reales? → $a_{-k}=a_k^*$.  
5️⃣ ¿Qué ocurre con $\omega_0$ si duplicas el período $T_0$? → $\omega_0$ se reduce a la mitad.

# 🧩 Problema 3.5 — Serie de Fourier de una **onda triangular impar**

> **Enunciado**  
> Sea $x(t)$ una **onda triangular impar**, de período $T_0$, amplitud pico $\pm A$, centrada en $t=0$.  
> Obtener su **Serie de Fourier** (forma seno–coseno), indicando la **simetría** y el **decaimiento** de los coeficientes.

---

## 🧠 Teoría breve

- **Imparidad**: si $x(t)$ es impar, la serie tiene **solo senos** (no hay cosenos ni componente DC):
  $$
  x(t)=\sum_{k=1}^{\infty} B_k\,\sin(k\omega_0 t),\qquad \omega_0=\frac{2\pi}{T_0}.
  $$
- **Suavidad**: señales más “suaves” (continuas y con derivadas por tramos) tienen coeficientes que decrecen más rápido.  
  La **triangular** decae como $1/k^2$ (más rápido que la cuadrada, $1/k$).

---

## 🧮 Método A (recomendado) — Derivar desde onda cuadrada e **integrar**

**Idea:** la **derivada** de una triangular impar ideal es una **onda cuadrada impar** (alterna entre $\pm$ constantes).

1) **Serie conocida de la cuadrada impar** (amplitud $\pm A$):
   $$
   q(t)=\frac{4A}{\pi}\!\left[\sin(\omega_0 t)+\frac{1}{3}\sin(3\omega_0 t)+\frac{1}{5}\sin(5\omega_0 t)+\cdots\right].
   $$

2) **Relación con la triangular**: si $x'(t)=q(t)$, entonces integrar “armónico a armónico” introduce un factor $\tfrac{1}{k\omega_0}$:
   $$
   x(t)=\sum_{\substack{k=1\\k\ \text{impar}}}^{\infty}
   \frac{4A}{\pi}\cdot\frac{1}{k\omega_0}\,\sin(k\omega_0 t).
   $$

3) **Sustituir $\omega_0=2\pi/T_0$** y ajustar la constante para que el pico sea exactamente $\pm A$:
   $$
   \boxed{
   x(t)=\sum_{\substack{k=1\\k\ \text{impar}}}^{\infty}
   \frac{8A}{\pi^2}\cdot\frac{1}{k^2}\,\sin(k\omega_0 t),
   \qquad \omega_0=\frac{2\pi}{T_0}.
   }
   $$

> **Conclusión del método A**:  
> - Solo aparecen **armónicos impares** ($k=1,3,5,\dots$).  
> - Los coeficientes **decaen como $1/k^2$**.

---

## 🔎 Comprobaciones rápidas

- **Imparidad** ⇒ $a_0=0$ y **sin cosenos** (solo términos seno).  
- **Decaimiento $1/k^2$** ⇒ reconstrucción más “suave” y convergencia rápida.  
- **Solo impares** ⇒ por la simetría de la forma triangular impar.

---

## 📊 Propiedades (tabla GitHub–friendly)

| Propiedad | Valor |
|:--|:--|
| Simetría | Impar → solo **senos** |
| Componente DC | $a_0=0$ |
| Coeficientes | $B_k=\dfrac{8A}{\pi^2 k^2}$ (solo $k$ **impares**) |
| Decaimiento | $B_k\sim \dfrac{1}{k^2}$ |
| Período | $T_0$ dado → $\omega_0=\dfrac{2\pi}{T_0}$ |

---

## ✅ Resultado final (Problema 3.5)

$$
\boxed{
x(t)=\sum_{\substack{k=1\\k\ \text{impar}}}^{\infty}
\frac{8A}{\pi^2 k^2}\,\sin(k\omega_0 t),\qquad
\omega_0=\frac{2\pi}{T_0}.
}
$$

---

## 🧩 Mini–test (auto–chequeo)

1) ¿Por qué no hay cosenos ni DC?  
   → Por **imparidad**: $a_0=0$ y solo aparecen senos.

2) ¿De dónde sale el $1/k^2$?  
   → De que la triangular es la **integral** de una cuadrada (que tiene $1/k$). Integrar añade un factor $1/(k\omega_0)$ a cada armónico.

3) ¿Por qué solo armónicos impares?  
   → Por la **simetría** de la triangular impar (antisimetría por semiperíodos).

4) ¿Qué pasa si desplazo la señal un tiempo $t_0$?  
   → Los módulos $|B_k|$ no cambian, pero aparece **fase**: $\sin(k\omega_0(t-t_0))$.

5) ¿Qué sucede si la señal tiene más suavidad (más derivadas continuas)?  
   → Los coeficientes decaen **aún más rápido** que $1/k^2$.


# 🧩 Problema 3.6 — Determinar si una señal es real, par o impar

> **Enunciado**  
> Dada una señal periódica $x(t)$, analizar si es **real**, **par** o **impar**,  
> y explicar cómo esas propiedades afectan a su **Serie de Fourier**.

---

## 🧠 Teoría esencial

1. **Señal real**  
   - $x(t) \in \mathbb{R}$ para todo $t$.  
   - En la Serie de Fourier exponencial:  
     $$x(t) = \sum_{k=-\infty}^{\infty} a_k e^{j k \omega_0 t},$$  
     la condición de realidad es  
     $$a_{-k} = a_k^\*.$$

2. **Señal par o impar**
   - Par: $x(t) = x(-t)$  → solo **cosenos**.  
   - Impar: $x(t) = -x(-t)$ → solo **senos**.  
   - En general (serie seno–coseno):  
     $$x(t) = a_0 + \sum_{k=1}^{\infty}[A_k \cos(k\omega_0 t) + B_k \sin(k\omega_0 t)].$$  
     Si es par → $B_k=0$.  
     Si es impar → $a_0=A_k=0$.

3. **Interpretación física**
   - Señal par → simetría vertical.  
   - Señal impar → simetría rotacional alrededor del origen.  
   - Señal real → amplitud sin parte imaginaria.

---

## 🧮 Caso A — $x(t) = \cos(4t) + \cos(6t)$

- $\cos$ es función par y real.  
- $x(-t)=x(t)$ → **par**.  
- $a_{-k}=a_k^\*$ → **real**.  

✅ **Conclusión:** señal **real y par**.  
**Serie de Fourier:** solo cosenos y DC.

---

## 🧮 Caso B — $x(t) = \sin(5t) - \sin(7t)$

- $\sin$ es función impar y real.  
- $x(-t) = -x(t)$ → **impar**.  

✅ **Conclusión:** señal **real e impar**.  
**Serie de Fourier:** solo senos, sin componente DC.

---

## 🧮 Caso C — $x(t) = \cos(2t) + \sin(3t)$

- Mezcla de par ($\cos$) e impar ($\sin$).  
- $x(-t) \neq \pm x(t)$ → **ni par ni impar**.  

✅ **Conclusión:** señal **real** pero sin simetría.

---

## 🧮 Caso D — $x(t) = e^{j3t}$

- Contiene parte imaginaria → **no es real**.  
- $a_{+3}=1$ y el conjugado $a_{-3}=0$.  

✅ **Conclusión:** señal **compleja**, ni par ni impar.

---

## 🧾 Resumen de propiedades

| Tipo de señal | Condición temporal | Forma espectral |
|:--|:--|:--|
| Real | $x(t)\in\mathbb{R}$ | $a_{-k}=a_k^\*$ |
| Par | $x(t)=x(-t)$ | $a_k=a_{-k}\in\mathbb{R}$ |
| Impar | $x(t)=-x(-t)$ | $a_k=-a_{-k}\in j\mathbb{R}$ |

---

## ✅ Conclusión general (Problema 3.6)

- **Real + Par** → sólo cosenos ($B_k=0$).  
- **Real + Impar** → sólo senos ($A_k=0$, $a_0=0$).  
- **No simétrica** → aparecen ambos términos.  

---

## 🧩 Mini-test tipo examen

1. ¿Qué condición garantiza que $x(t)$ sea real?  
   → $a_{-k}=a_k^\*.$  

2. ¿Qué ocurre si $x(t)$ es par?  
   → $B_k=0$ (solo cosenos).  

3. ¿Y si $x(t)$ es impar?  
   → $a_0=A_k=0$ (solo senos).  

4. ¿Qué pasa si $x(t)$ es compleja?  
   → El espectro no cumple simetría hermítica.  

5. ¿Cómo se detecta la imparidad en el tiempo?  
   → Si $x(-t)=-x(t)$.

# 🧩 Problema 3.8 — Construcción de señales con propiedades específicas

> **Enunciado**  
> A partir de un conjunto de coeficientes $a_k$, construir una señal periódica $x(t)$  
> que cumpla ciertas **propiedades** (real, par, impar, compleja).  
> Indicar cómo deben ser los coeficientes $a_k$ en cada caso.

---

## 🧠 Teoría breve

La Serie de Fourier exponencial se escribe como

$$x(t)=\sum_{k=-\infty}^{\infty} a_k e^{j k \omega_0 t}.$$

El **aspecto del espectro** determina la naturaleza de la señal:

| Tipo de señal | Condición espectral | Ejemplo |
|:--|:--|:--|
| Real | $a_{-k}=a_k^\*$ | $x(t)=2\cos(\omega_0 t)$ |
| Par | $a_k=a_{-k}\in\mathbb{R}$ | $x(t)=A_1\cos(\omega_0 t)+A_2\cos(2\omega_0 t)$ |
| Impar | $a_k=-a_{-k}\in j\mathbb{R}$ | $x(t)=B_1\sin(\omega_0 t)+B_2\sin(3\omega_0 t)$ |
| Compleja | sin relación entre $a_{+k}$ y $a_{-k}$ | $x(t)=e^{j\omega_0 t}$ |

---

## 🧮 Ejemplo 1 — Señal real y par

Queremos una señal real **par**.  
Condición: $a_{-k}=a_k^\*$ y además $a_k$ real.

Elijamos:

| $k$ | $a_k$ |
|:--:|:--:|
| 0 | 1 |
| ±1 | 0.5 |
| resto | 0 |

Entonces:

$$x(t)=a_0+a_1 e^{j\omega_0 t}+a_{-1}e^{-j\omega_0 t}
=1+0.5(e^{j\omega_0 t}+e^{-j\omega_0 t})
=1+\cos(\omega_0 t).$$

✅ **Señal real y par** (solo cosenos).

---

## 🧮 Ejemplo 2 — Señal real e impar

Condición: $a_{-k}=-a_k^\*$ y $a_k$ puramente imaginarios.

Tomemos:

| $k$ | $a_k$ |
|:--:|:--:|
| +1 | $j/2$ |
| −1 | $-j/2$ |

Entonces:

$$x(t)=a_1 e^{j\omega_0 t}+a_{-1} e^{-j\omega_0 t}
=\frac{j}{2}(e^{j\omega_0 t}-e^{-j\omega_0 t})
=\sin(\omega_0 t).$$

✅ **Señal real e impar.**

---

## 🧮 Ejemplo 3 — Señal compleja

Elige $a_{+1}=1$, $a_{-1}=0$:

$$x(t)=e^{j\omega_0 t}.$$

No cumple $a_{-1}=a_1^\*$ → **compleja pura**.  
Su parte real es $\cos(\omega_0 t)$ (par) y su parte imaginaria $\sin(\omega_0 t)$ (impar).

---

## 🧾 Resumen (tabla de síntesis)

| Propiedad de la señal | Condición sobre $a_k$ | Forma temporal |
|:--|:--|:--|
| Real | $a_{-k}=a_k^\*$ | Parte real de exponenciales |
| Par | $a_k=a_{-k}\in\mathbb{R}$ | Solo cosenos |
| Impar | $a_k=-a_{-k}\in j\mathbb{R}$ | Solo senos |
| Compleja | Sin relación | Incluye partes reales e imaginarias |

---

## ✅ Conclusión (Problema 3.8)

El espectro **determina la simetría** de la señal:  
- Si el espectro es **hermítico** → señal **real**.  
- Si además es **simétrico** → señal **par**.  
- Si es **antisimétrico** → señal **impar**.  
- Si no cumple nada de eso → señal **compleja**.

---

## 🧩 Mini-test rápido

1️⃣ ¿Qué relación entre $a_k$ y $a_{-k}$ define una señal real?  
→ $a_{-k}=a_k^\*$.

2️⃣ Si $a_k=a_{-k}$ y todos son reales, ¿qué tipo de señal es?  
→ Par.

3️⃣ Si $a_k=-a_{-k}$ y todos son imaginarios, ¿qué tipo de señal es?  
→ Impar.

4️⃣ ¿Qué ocurre si solo existe $a_3=1$?  
→ Señal compleja pura $x(t)=e^{j3\omega_0 t}$.

5️⃣ ¿Qué representa la simetría del espectro respecto a $k=0$?  
→ La **simetría temporal** de la señal en el dominio del tiempo.

# 🧩 Problema 3.11 — DTFS de una coseno discreta y recuperación de A, B, C

> **Enunciado**  
> 1) Dada una señal periódica discreta  
> $$x[n]=A\cos(Bn+C),\qquad x[n+N_0]=x[n],$$  
> obtener su **DTFS** (coeficientes $X[k]$).  
> 2) Problema inverso: dados unos coeficientes $X[k]$ no nulos en dos líneas conjugadas, **recuperar $A$, $B$ (o $m_0$) y $C$**.

---

## 🧠 Teoría breve (DTFS)

Para una señal **periódica** discreta de **período $N_0$**:

- **Síntesis** (reconstrucción):
  $$
  x[n]=\sum_{k=0}^{N_0-1} X[k]\;e^{j\frac{2\pi}{N_0}kn}.
  $$
- **Análisis** (cálculo de coeficientes):
  $$
  X[k]=\frac{1}{N_0}\sum_{n=0}^{N_0-1} x[n]\;e^{-j\frac{2\pi}{N_0}kn}.
  $$
- Señal **real** $\Rightarrow$ **simetría hermítica**:  
  $$X[N_0-k]=X[k]^\*.$$

> Una **coseno discreta pura** de la forma $x[n]=A\cos\!\big(\tfrac{2\pi m_0}{N_0}n+\phi\big)$ tiene **solo dos líneas** en el espectro: $k=m_0$ y $k=N_0-m_0$.

---

## 🧮 Paso 1 — DTFS de $x[n]=A\cos\!\left(\dfrac{2\pi m_0}{N_0}n+\phi\right)$

Usamos Euler:
$$
\cos\theta=\frac{e^{j\theta}+e^{-j\theta}}{2}.
$$

Entonces
$$
x[n]=\frac{A}{2}\,e^{j\left(\frac{2\pi m_0}{N_0}n+\phi\right)}
       +\frac{A}{2}\,e^{-j\left(\frac{2\pi m_0}{N_0}n+\phi\right)}.
$$

Comparamos con la base DTFS $e^{j\frac{2\pi}{N_0}kn}$:

- El primer término coincide con $k=+m_0$ y coeficiente **constante** (no depende de $n$):
  $$
  X[m_0]=\frac{A}{2}\,e^{j\phi}.
  $$
- El segundo término coincide con $k=N_0-m_0$ (porque $e^{-j\frac{2\pi}{N_0}m_0 n}=e^{j\frac{2\pi}{N_0}(N_0-m_0)n}$):
  $$
  X[N_0-m_0]=\frac{A}{2}\,e^{-j\phi}.
  $$

Y todos los demás $X[k]=0$.

**Resumen (resultado general):**
$$
\boxed{
X[m_0]=\frac{A}{2}\,e^{j\phi},\quad
X[N_0-m_0]=\frac{A}{2}\,e^{-j\phi},\quad
\text{resto } X[k]=0.}
$$

> **Nota:** $B=\dfrac{2\pi m_0}{N_0}$ (frecuencia discreta).  
> Si $m_0$ y $N_0$ comparten divisor, puede existir un período fundamental menor (reducir fracción).

---

## 🧪 Ejemplo directo (cálculo de $X[k]$)

**Datos**: $N_0=8$,  
$$x[n]=3\cos\!\left(\frac{2\pi\cdot 2}{8}\,n+\frac{\pi}{4}\right)=3\cos\!\left(\frac{\pi}{2}\,n+\frac{\pi}{4}\right).$$

Aquí $A=3$, $m_0=2$, $\phi=\pi/4$.

Aplicando la fórmula:
$$
X[2]=\frac{A}{2}e^{j\phi}=\frac{3}{2}e^{j\pi/4},\qquad
X[8-2]=X[6]=\frac{3}{2}e^{-j\pi/4},
$$
y $X[k]=0$ para $k\notin\{2,6\}$.

**Chequeo de realidad:** $X[6]=X[2]^\*$ ✔️

---

## 🔁 Paso 2 — Problema inverso (recuperar $A$, $m_0$, $\phi$)

**Datos**: Para una señal real y periódica $N_0$-periódica, se observan solo dos líneas no nulas y conjugadas:
$$
X[k_1]=R\,e^{j\alpha},\qquad
X[k_2]=R\,e^{-j\alpha},\qquad k_2=N_0-k_1.
$$

**Recuperación:**
- $m_0=k_1$ (o $k_2=N_0-k_1$).
- $A=2R$.
- $\phi=\alpha$ (fase de la línea en $k_1$).
- La señal es:
  $$
  x[n]=A\cos\!\left(\frac{2\pi m_0}{N_0}n+\phi\right).
  $$

### Ejemplo inverso
**Datos espectrales:** $N_0=12$, $X[3]=2e^{-j\pi/3}$ y $X[9]=2e^{+j\pi/3}$ (resto 0).  
**Cálculo:**
- $m_0=3$ (frecuencia $B=\tfrac{2\pi\cdot3}{12}=\tfrac{\pi}{2}$).
- $R=2$ ⇒ $A=2R=4$.
- Fase $\phi=\angle X[3]=-\pi/3$.
- **Resultado**:
  $$
  x[n]=4\cos\!\left(\frac{\pi}{2}n-\frac{\pi}{3}\right).
  $$

**Verificación rápida (síntesis DTFS):**
- Esperamos $X[3]=\tfrac{A}{2}e^{j\phi}=2e^{-j\pi/3}$ ✔️  
- Y $X[12-3]=X[9]=2e^{+j\pi/3}$ ✔️

---

## 🧾 Casos especiales y advertencias

1. **$m_0=0$ (DC pura):**  
   $x[n]=A\cos(\phi)=\text{constante} \Rightarrow X[0]=\text{constante},$ resto $0$.
2. **$m_0=N_0/2$ (si $N_0$ par):**  
   $x[n]=A\cos(\pi n+\phi)=A(-1)^n\cos\phi - A(-1)^n\sin\phi\,\tan(\cdots)$  
   En realidad, aquí la exponencial positiva y negativa coinciden en la **misma** bin (Nyquist).  
   Resultado típico: **una sola línea real** (signo según $\phi$).
3. **Fase desconocida por desplazamiento temporal:**  
   Un **desplazamiento** $n_0$ en el tiempo introduce **fase lineal**:  
   $x[n-n_0] \leftrightarrow X[k]\,e^{-j\frac{2\pi}{N_0}k n_0}$.  
   Si ves fase, puede ser por **retardo**.

---

## ✅ Resumen del Problema 3.11

- Si $x[n]=A\cos\!\big(\tfrac{2\pi m_0}{N_0}n+\phi\big)$, entonces:
  $$
  \boxed{
  X[m_0]=\frac{A}{2}e^{j\phi},\quad
  X[N_0-m_0]=\frac{A}{2}e^{-j\phi},\quad
  \text{resto } X[k]=0. }
  $$
- Inverso: si el espectro tiene **dos líneas conjugadas** en $k=m_0$ y $k=N_0-m_0$ con módulo $R$,  
  entonces $A=2R$ y $\phi=\angle X[m_0]$.

---

1) Si $x[n]=5\cos\!\big(\tfrac{2\pi\cdot4}{16}n+\tfrac{\pi}{6}\big)$, ¿qué $X[k]$ son no nulos?  
   **→** $X[4]=2.5\,e^{j\pi/6}$ y $X[12]=2.5\,e^{-j\pi/6}$.

2) Dado $N_0=10$, $X[2]=1.8\,e^{j\pi/5}$ y $X[8]=1.8\,e^{-j\pi/5}$, construye $x[n]$.  
   **→** $A=3.6$, $m_0=2$, $\phi=\pi/5$ ⇒ $x[n]=3.6\cos\!\big(\tfrac{2\pi\cdot2}{10}n+\tfrac{\pi}{5}\big)$.

3) ¿Qué condición garantiza que $x[n]$ es real?  
   **→** $X[N_0-k]=X[k]^\*$.

   **→** Cuando $m_0=N_0/2$ (si $N_0$ es par).

# 🧩 Problema 3.12 — Convolución periódica y producto en el tiempo (Serie de Fourier)

> **Objetivo**  
> 1) Definir la **convolución periódica** de dos señales periódicas $x(t)$ e $h(t)$ de período $T_0$.  
> 2) Relacionar la convolución en el tiempo con los **coeficientes de la Serie de Fourier (FS)**.  
> 3) Relacionar el **producto en el tiempo** $y(t)=x(t)\,h(t)$ con los coeficientes de la FS.  
> 4) Resolver ejemplos numéricos con pasos intermedios.

---

## 🧠 Teoría breve

Sea
$$
x(t)=\sum_{k=-\infty}^{\infty} a_k\,e^{jk\omega_0 t},\qquad
h(t)=\sum_{k=-\infty}^{\infty} b_k\,e^{jk\omega_0 t},\qquad
\omega_0=\frac{2\pi}{T_0}.
$$

### 1) Convolución **periódica** (sobre un período)
Definimos
$$
(y)(t)=(x *_T h)(t)\;=\;\int_{t_0}^{t_0+T_0} x(\tau)\,h(t-\tau)\,d\tau,
$$
con cualquier $t_0$ (el resultado no depende de $t_0$ si $x$ y $h$ son $T_0$-periódicas).

**Resultado clave (en FS):**
si $y(t)=x *_T h$, entonces sus coeficientes de FS son
$$
\boxed{\,y_k=T_0\,a_k\,b_k\,}\qquad(\text{escalado punto a punto por $T_0$}).
$$

> Intuición: en **convolución periódica**, **cada armónico $k$ se escala** por el producto de los coeficientes en ese mismo $k$ (y por $T_0$).

---

### 2) **Producto** en el tiempo
Si $y(t)=x(t)\,h(t)$, entonces los coeficientes de FS se obtienen por **convolución discreta** en el índice $k$:
$$
\boxed{\,y_k=\sum_{m=-\infty}^{\infty} a_m\,b_{k-m}\,}\qquad(\text{convolución de espectros}).
$$

> Intuición: multiplicar señales en tiempo mezcla armónicos → **aparecen sumas de frecuencias** $k=m+(k-m)$.

---

## 🧮 Parte A — Ejemplo de **convolución periódica**

Tomemos
- $x(t)=\cos(\omega_0 t)$, cuya FS es $a_{+1}=a_{-1}=1/2$ (resto cero).
- $h(t)=1+\cos(\omega_0 t)$, cuya FS es $b_0=1$, $b_{+1}=b_{-1}=1/2$ (resto cero).

**Queremos:** $y(t)=(x *_T h)(t)$ y sus coeficientes $y_k$.

### Paso 1 — Coeficientes de entrada
- $x(t)$: $a_{+1}=1/2$, $a_{-1}=1/2$, demás $a_k=0$.
- $h(t)$: $b_0=1$, $b_{+1}=1/2$, $b_{-1}=1/2$, demás $b_k=0$.

### Paso 2 — Fórmula de convolución periódica
$$
y_k=T_0\,a_k\,b_k.
$$

### Paso 3 — Cálculo armónico a armónico
- $k=0$: $a_0=0$ y $b_0=1$ ⇒ $y_0=T_0(0)(1)=0$.
- $k=+1$: $a_{+1}=\tfrac12$, $b_{+1}=\tfrac12$ ⇒ $y_{+1}=T_0\cdot\tfrac12\cdot\tfrac12=\tfrac{T_0}{4}$.
- $k=-1$: $a_{-1}=\tfrac12$, $b_{-1}=\tfrac12$ ⇒ $y_{-1}=\tfrac{T_0}{4}$.
- Resto: $y_k=0$ (al menos uno de $a_k$ o $b_k$ es cero).

### Paso 4 — Reconstrucción de $y(t)$
$$
y(t)=\sum_k y_k e^{jk\omega_0 t}
=\frac{T_0}{4}\,e^{j\omega_0 t}+\frac{T_0}{4}\,e^{-j\omega_0 t}
=\frac{T_0}{2}\cos(\omega_0 t).
$$

**Conclusión:** la **convolución periódica** aquí **escala la amplitud** de la frecuencia común $\omega_0$ por un factor $T_0/2$ y elimina el resto.

> Nota: el factor $T_0$ es inherente a la definición de **convolución periódica**. En exámenes, lo más habitual es que **aparezcan exactamente estos factores**.

---

## 🧮 Parte B — Ejemplo de **producto en el tiempo**

Ahora sea
$$
x(t)=\cos(\omega_0 t),\qquad h(t)=\cos(\omega_0 t).
$$
Sabemos que
$$
x(t)\,h(t)=\cos(\omega_0 t)\cos(\omega_0 t)
=\tfrac{1}{2}\big(1+\cos(2\omega_0 t)\big).
$$

Vamos a **reproducir ese resultado** usando coeficientes de FS y la fórmula
$$
y_k=\sum_{m=-\infty}^{\infty} a_m\,b_{k-m}.
$$

### Paso 1 — Coeficientes de entrada
Para cada coseno: $a_{+1}=a_{-1}=1/2$, resto $0$; idem $b_k$.

### Paso 2 — Convolución de coeficientes
- $y_0=\sum_m a_m b_{0-m}=a_{+1}b_{-1}+a_{-1}b_{+1}
=\tfrac12\cdot\tfrac12+\tfrac12\cdot\tfrac12=\tfrac12.$
- $y_{+2}=\sum_m a_m b_{2-m}=a_{+1}b_{+1}=\tfrac12\cdot\tfrac12=\tfrac14.$
- $y_{-2}=\sum_m a_m b_{-2-m}=a_{-1}b_{-1}=\tfrac12\cdot\tfrac12=\tfrac14.$
- Otros $k$: cero (no hay combinaciones que sumen a $|k|\neq 0,2$).

### Paso 3 — Reconstrucción
$$
y(t)=y_0 + y_{+2}e^{j2\omega_0 t}+y_{-2}e^{-j2\omega_0 t}
=\tfrac12+\tfrac14 e^{j2\omega_0 t}+\tfrac14 e^{-j2\omega_0 t}
=\tfrac12+\tfrac12\cos(2\omega_0 t).
$$

**Coincide** con la identidad trigonométrica.  
**Conclusión:** el **producto en tiempo** genera **sumas/restas de frecuencias** (convolución de índices $k$ en los coeficientes).

---

## 🧾 Resumen (para examen)

- **Convolución periódica**:  
  $$y(t)=x *_T h \quad\Longrightarrow\quad \boxed{y_k=T_0\,a_k\,b_k}.$$
  Cada armónico $k$ se **escala** (no se mezclan $k$ diferentes).

- **Producto en tiempo**:  
  $$y(t)=x(t)\,h(t)\quad\Longrightarrow\quad \boxed{y_k=\sum_m a_m\,b_{k-m}}.$$
  Los armónicos se **combinan** (aparecen nuevas frecuencias $k$ por suma de índices).

---

## 🧩 Mini-test (auto-chequeo)

1) Si $x(t)$ tiene solo $a_{+1}=a_{-1}=1/2$ y $h(t)$ solo $b_{+2}=b_{-2}=1/2$,  
   ¿qué $y_k$ aparecen para $y=x\cdot h$?  
   **Pista:** $k=\pm3$ y $k=\pm1$ (sumas y restas de índices).

2) En $y=x *_T h$, si $x$ y $h$ solo tienen armónico $k=3$,  
   ¿qué vale $y_3$? ¿Aparecen otros $y_k$?  
   **Sol.:** $y_3=T_0 a_3 b_3$, el resto 0.

3) Si $x$ es **par** (solo cosenos) y $h$ es **par**, ¿qué simetría tendrá $y=x\cdot h$?  
   **Sol.:** par (producto de pares es par) → solo cosenos en $y$.

4) Si $x$ es real y $h$ es real, ¿qué condición cumplen $y_k$?  
   **Sol.:** $y_{-k}=y_k^\*$ (simetría hermítica).

5) ¿Qué diferencia clave hay entre $x *_T h$ y $x\cdot h$ en el dominio $k$?  
   **Sol.:** en $x *_T h$ hay **producto punto a punto** con factor $T_0$;  
   en $x\cdot h$ hay **convolución discreta** de coeficientes.

# 🧩 Problema 3.14 — Salida de un sistema LTI dado $H(j\omega)$ y $a_k$

> **Objetivo**  
> Dada una entrada periódica 
> $$x(t)=\sum_{k=-\infty}^{\infty} a_k\,e^{jk\omega_0 t}$$
> que entra a un **sistema LTI continuo** con respuesta frecuencial $H(j\omega)$, obtener la salida
> $$y(t)=\sum_{k=-\infty}^{\infty} y_k\,e^{jk\omega_0 t}$$
> y expresar **$y_k$ en función de $a_k$ y $H(j\omega)$**.  
> Resuelve ejemplos numéricos típicos (filtros PB/PA y retardo puro).

---

## 🧠 Teoría breve

Un sistema **LTI** no cambia la forma exponencial $e^{j\omega t}$: solo **escala** y **gira** (amplitud y fase).  
Si la entrada es periódica con armónicos $k\omega_0$, entonces **cada armónico** se multiplica por el valor del sistema en esa frecuencia:

$$
\boxed{\,y_k = H\!\big(jk\omega_0\big)\; a_k\,}
$$

Por tanto,
$$
y(t)=\sum_{k=-\infty}^{\infty} \underbrace{H(jk\omega_0)a_k}_{y_k}\,e^{jk\omega_0 t}.
$$

**Lectura física**:  
- $|H(jk\omega_0)|$ ajusta la **amplitud** del armónico $k$.  
- $\angle H(jk\omega_0)$ suma **fase** al armónico $k$.

---

## ✅ Procedimiento (receta de examen)

1. **Escribe/lee** la FS de la entrada: $x(t)=\sum a_k e^{jk\omega_0 t}$.  
2. **Evalúa** el sistema en cada armónico: $H(jk\omega_0)$.  
3. **Multiplica** coeficiente a coeficiente: $y_k=H(jk\omega_0)a_k$.  
4. **Reconstruye**: $y(t)=\sum y_k e^{jk\omega_0 t}$.  
5. **(Opcional)**: pasa a forma seno–coseno si lo piden.

---

## 🧮 Ejemplo 1 — Filtro **Paso-Bajo ideal** (PB) con onda cuadrada impar

**Entrada**: onda cuadrada impar (amplitud pico $\pm A$, período $T_0$)
$$
x(t)=\frac{4A}{\pi}\left[\sin(\omega_0 t)+\frac{1}{3}\sin(3\omega_0 t)+\frac{1}{5}\sin(5\omega_0 t)+\cdots\right].
$$
En forma exponencial (solo impares $k=\pm1,\pm3,\dots$), los $a_k$ cumplen
$$
|a_k|=\frac{2A}{\pi}\,\frac{1}{|k|},\quad \angle a_k =
\begin{cases}
-\frac{\pi}{2}, & k>0 \\
+\frac{\pi}{2}, & k<0
\end{cases}
\quad\text{(por ser senos)}.
$$

**Sistema**: PB ideal de corte en $\omega_c=2\omega_0$:
$$
H(j\omega)=
\begin{cases}
1, & |\omega|\le \omega_c \\
0, & \text{si no}
\end{cases}
\quad\Longrightarrow\quad
H(jk\omega_0)=
\begin{cases}
1, & |k|\le 2 \\
0, & \text{si no}
\end{cases}
$$

**Paso 1. Armónicos que pasan**: como solo **impares** existen en $x(t)$, con el PB quedan **$k=\pm1$** (pues $|1|\le2$ y $|3|>2$ se anula, etc.).  

**Paso 2. Coeficientes de salida**: $y_k=H(jk\omega_0)a_k$:
- $y_{\pm1}=a_{\pm1}$,
- $y_{\pm3}=0$, $y_{\pm5}=0$, …

**Reconstrucción**:
$$
y(t)=\frac{4A}{\pi}\sin(\omega_0 t)
$$
(¡solo el **primer armónico**!).  
**Conclusión**: el PB “suaviza” quedándose con el fundamental.

---

## 🧮 Ejemplo 2 — Filtro **Paso-Alto ideal** (PA) con la misma entrada

**Sistema**: PA ideal de corte en $\omega_c=2\omega_0$:
$$
H(j\omega)=
\begin{cases}
0, & |\omega|<\omega_c \\
1, & \text{si no}
\end{cases}
\quad\Longrightarrow\quad
H(jk\omega_0)=
\begin{cases}
0, & |k|<2 \\
1, & \text{si no}
\end{cases}
$$

**Armónicos que pasan**: $k=\pm3,\pm5,\dots$ (se eliminan $\pm1$).  

**Coeficientes de salida**: $y_{\pm1}=0$, y $y_{\pm(2m+1)}=a_{\pm(2m+1)}$ para $m\ge1$.

**Reconstrucción**:
$$
y(t)=\frac{4A}{\pi}\left[\frac{1}{3}\sin(3\omega_0 t)+\frac{1}{5}\sin(5\omega_0 t)+\cdots\right].
$$

**Conclusión**: el PA elimina el fundamental y deja los **bordes** (armónicos altos), por eso la forma es más “cuadrada/afilada”.

---

## 🧮 Ejemplo 3 — **Retardo puro**: $H(j\omega)=e^{-j\omega\tau}$

**Sistema**: $H(j\omega)=e^{-j\omega\tau}$ (LTI estable y causal para $\tau\ge0$).  
**Regla**:
$$
H(jk\omega_0)=e^{-jk\omega_0\tau}\quad\Longrightarrow\quad
\boxed{\,y_k=a_k\,e^{-jk\omega_0\tau}\,}
$$

**Interpretación**: no cambia ninguna amplitud, **solo añade fase lineal** → la salida es la **misma señal retrasada**:
$$
y(t)=x(t-\tau).
$$

**Ejemplo rápido**: si $x(t)$ es la cuadrada anterior, entonces $y(t)$ es la misma cuadrada **desplazada** $\tau$ en el tiempo (misma serie, fases rotadas por $-k\omega_0\tau$).

---

## 🧾 Observaciones útiles (para examen)

- Si $x(t)$ es **real**, $a_{-k}=a_k^\*$; si el sistema es fisicamente realizable y real, también suele cumplirse $H(-j\omega)=H(j\omega)^\*$ → entonces **$y(t)$ es real** y $y_{-k}=y_k^\*$.  
- **Causalidad** no se ve directamente en $H(j\omega)$ (hace falta $H(s)$ y ROC), pero **retardos** $e^{-j\omega\tau}$ con $\tau\ge0$ son causales.  
- **Estabilidad BIBO** en LTI-CT se verifica con $h(t)\in L^1$; en frecuencia, $H(j\omega)$ acotada y físicamente realizable.  
- Un **filtro pasa-todo** tendría $|H(j\omega)|=1$ y fase no trivial: modifica **fase**, no **magnitud**.

---

## ✅ Resumen (fórmula clave)

- **Salida armónica a armónica**:
  $$
  \boxed{\,y_k=H(jk\omega_0)\,a_k\,}
  \qquad\Rightarrow\qquad
  y(t)=\sum_k H(jk\omega_0)a_k\,e^{jk\omega_0 t}.
  $$

- Casos típicos:
  - PB/PA/BB: anulan o atenúan ciertos $k$ (magnitud).  
  - Retardo puro: añade $-k\omega_0\tau$ a la fase (no cambia magnitud).  
  - Filtros reales → salida real si entrada es real.

## 🧩 Mini-test rápido

1) Si $x(t)$ tiene solo $a_{\pm2}\neq0$ y $H(j\omega)$ es PB con corte $\omega_c=1.5\,\omega_0$, ¿qué sale?  
   **→** Como $|2|>1.5$, $H(j2\omega_0)=0$ ⇒ $y(t)=0$.

2) Para retardo puro $H(j\omega)=e^{-j\omega\tau}$, ¿cómo se relacionan $y_k$ y $a_k$?  
   **→** $y_k=a_k e^{-jk\omega_0\tau}$.

3) Si $H(j\omega)$ anula $k=\pm1$ y deja $k=\pm3$, ¿qué tipo de filtro es cerca de $\omega_0$?  
   **→** Paso-alto (elimina el fundamental).

4) Si $|H(jk\omega_0)|=1$ para todo $k$ y $\angle H(jk\omega_0)=-k\theta$, ¿qué hace el sistema?  
   **→** Pasa-todo con retardo de grupo constante $\theta/\omega_0$ (fase lineal).

5) ¿Cómo reconstruyes $y(t)$ si conoces todos los $y_k$?  
   **→** $y(t)=\sum_k y_k e^{jk\omega_0 t}$.
