### 1Ô∏è‚É£ **Serie de Fourier (FS)**  
<details>
<summary>üí° ¬øQu√© representa la Serie de Fourier?</summary>

**Teor√≠a breve:**

### 4Ô∏è‚É£ **Sistemas LTI y Convoluci√≥n**  
<details>
<summary>üí° ¬øC√≥mo se calcula la salida de un sistema LTI?</summary>

**Respuesta:**

$$

**Propiedades importantes:**
- Conmutativa ‚Üí $x*h = h*x$  
- Asociativa ‚Üí $(x*h_1)*h_2 = x*(h_1*h_2)$  
- Distributiva ‚Üí $x*(h_1+h_2)=x*h_1+x*h_2$

### 6Ô∏è‚É£ **Muestreo y aliasing**  
<details>
<summary>üí° ¬øQu√© es el muestreo y c√≥mo se evita el aliasing?</summary>

**Respuesta:**  
El muestreo consiste en tomar muestras de una se√±al continua cada $T_s$ segundos:
$$

$$
La frecuencia de muestreo es:
$$

$$

Para evitar **aliasing**, se debe cumplir el **Teorema de Nyquist**:
$$
f_s > 2f_{max}
$$

---

**Teor√≠a breve:**  
El aliasing ocurre cuando las copias del espectro se superponen.  
Si $f_s$ es demasiado bajo, las frecuencias se mezclan y la se√±al original no puede recuperarse.

**Conceptos previos:**

<summary>üìò Problema 1 ‚Äî Propiedades de un sistema continuo</summary>

**Enunciado**  
Sea el sistema de tiempo continuo definido por:
$$
y(t)=\int_{t-T}^{t}x(\tau)\,d\tau, \quad T>0
$$

Analizar si el sistema presenta las propiedades de **estabilidad**, **invarianza temporal** y **causalidad**.

---

### 1Ô∏è‚É£ Estabilidad
Si la entrada est√° acotada, $|x(t)|\le B$, se cumple:
$$
|y(t)|=\left|\int_{t-T}^{t}x(\tau)\,d\tau\right|
\le \int_{t-T}^{t}|x(\tau)|\,d\tau
\le B\int_{t-T}^{t}d\tau = BT < \infty
$$
El sistema **no amplifica indefinidamente** las se√±ales acotadas.

**Conclusi√≥n:** **BIBO estable** (Bounded Input ‚Üí Bounded Output).

---

### 2Ô∏è‚É£ Invarianza en el tiempo
Si se desplaza la entrada:
$$
x'(t)=x(t-t_0)
\Rightarrow
y'(t)=\int_{t-T}^{t}x(\tau-t_0)\,d\tau
=\int_{t-T-t_0}^{t-t_0}x(\sigma)\,d\sigma
= y(t-t_0)
$$

**Conclusi√≥n:** el sistema **es invariante**.  
El comportamiento no cambia en el tiempo: un desplazamiento en la entrada produce el mismo desplazamiento en la salida.

---

### 3Ô∏è‚É£ Causalidad
Para $x(t)=\delta(t)$,
$$
h(t)=\int_{t-T}^{t}\delta(\tau)\,d\tau=u(t)-u(t-T)
$$
Como $h(t)=0$ para $t<0$, la salida solo depende de valores **pasados o presentes** de la entrada.

**Conclusi√≥n:** el sistema es **causal**.

---

### üß† Teor√≠a
Este sistema **integra la se√±al de entrada en una ventana finita** de duraci√≥n $T$.  
- Act√∫a como un **promediador m√≥vil**: la salida en cada instante $t$ representa la energ√≠a promedio de la se√±al en el intervalo $[t-T,\,t]$.  
- Es **LTI** (lineal e invariante en el tiempo) porque cumple ambas propiedades y su respuesta impulsional es fija.  
- Este tipo de sistema suaviza la se√±al, reduciendo ruido o variaciones r√°pidas (similar a un filtro paso bajo en frecuencia).

## Problema 1 (3 puntos) ‚Äî Sistema DT con ventana finita y factor cosenoidal

Sea el sistema de tiempo discreto:
$$
y[n]=\frac{1}{N+1}\sum_{k=n-N}^{n} x[k]\;\cos\!\left(\frac{\pi n}{2}\right), 
\qquad N\in\mathbb{Z}^+,\; N\ge 2.
$$

Se pide analizar: **(a)** causalidad, **(b)** estabilidad BIBO, **(c)** linealidad (aditividad y homogeneidad), **(d)** invariancia temporal; y **(2)** obtener la respuesta al impulso \(h[n]\) y representarla para \(N=3\).

---

<details>
<summary>1.a) Causalidad</summary>

**An√°lisis.**  
Para un √≠ndice fijo \(n_0\),
$$
y[n_0]=\frac{1}{N+1}\sum_{k=n_0-N}^{n_0} x[k]\;\cos\!\left(\frac{\pi n_0}{2}\right).
$$
La suma usa √∫nicamente valores de \(x[k]\) con \(k\le n_0\) (pasado y presente).

**Conclusi√≥n.** El sistema es **causal**.
</details>

---

<details>
<summary>1.b) Estabilidad BIBO</summary>

**Hip√≥tesis.** Si \(|x[n]|\le B\) para todo \(n\), entonces
$$
|y[n]|
\le \frac{1}{N+1}\sum_{k=n-N}^{n} |x[k]|\; \bigl|\cos\!\left(\tfrac{\pi n}{2}\right)\bigr|
\le \frac{1}{N+1}\sum_{k=n-N}^{n} B
= B.
$$

**Conclusi√≥n.** Toda entrada acotada produce salida acotada: **BIBO estable**.
</details>

---

<details>
<summary>1.c) Linealidad (aditividad y homogeneidad)</summary>

**Aditividad.** Para \(x[n]=x_1[n]+x_2[n]\),
$$
y[n]=\frac{1}{N+1}\sum_{k=n-N}^{n} \bigl(x_1[k]+x_2[k]\bigr)\cos\!\left(\tfrac{\pi n}{2}\right)
= y_1[n]+y_2[n].
$$

**Homogeneidad.** Para \(x'[n]=\alpha x[n]\),
$$
y'[n]=\frac{1}{N+1}\sum_{k=n-N}^{n} \alpha x[k]\cos\!\left(\tfrac{\pi n}{2}\right)=\alpha\,y[n].
$$

**Conclusi√≥n.** El sistema es **lineal**.
</details>

---

<details>
<summary>1.d) Invariancia temporal</summary>

**Prueba.** Con \(x'(n)=x[n-n_0]\),
$$
y'[n]=\frac{1}{N+1}\sum_{k=n-N}^{n} x[k-n_0]\cos\!\left(\tfrac{\pi n}{2}\right).
$$
En cambio,
$$
y[n-n_0]=\frac{1}{N+1}\sum_{k=n-n_0-N}^{n-n_0} x[k]\;\underbrace{\cos\!\left(\tfrac{\pi (n-n_0)}{2}\right)}_{\neq \cos(\pi n/2)\, \text{en general}}.
$$
El factor \(\cos(\tfrac{\pi n}{2})\) **depende expl√≠citamente de \(n\)**, por lo que, en general, \(y'[n]\ne y[n-n_0]\).

**Conclusi√≥n.** **No invariante en el tiempo** (no TI).
</details>

---

<details>
<summary>2) Respuesta al impulso \(h[n]\) y caso \(N=3\)</summary>

**Definici√≥n.** Para \(x[n]=\delta[n]\),
$$
h[n]=\frac{1}{N+1}\sum_{k=n-N}^{n}\delta[k]\;\cos\!\left(\tfrac{\pi n}{2}\right)
=\frac{1}{N+1}\cos\!\left(\tfrac{\pi n}{2}\right)\,\bigl(u[n]-u[n-N-1]\bigr).
$$

**Particularizaci√≥n \(N=3\).**
$$
h[n]=\frac{1}{4}\cos\!\left(\tfrac{\pi n}{2}\right)\,\bigl(u[n]-u[n-4]\bigr),
$$
de modo que
\[
h[0]=\tfrac14,\quad
h[1]=0,\quad
h[2]=-\tfrac14,\quad
h[3]=0,\quad
h[n]=0\ \text{si}\ n<0\ \text{o}\ n\ge 4.
\]

**Descripci√≥n gr√°fica (N=3).** Secuencia finita de longitud 4 con muestras:
\((n,h[n])=\{(0,\tfrac14),\ (1,0),\ (2,-\tfrac14),\ (3,0)\}\).
</details>

---

### Teor√≠a (conexiones y lectura r√°pida)
- El sistema **promedia** \(N\!+\!1\) muestras pasadas y presentes de \(x[k]\) (ventana rectangular) y las **modula** por \(\cos(\pi n/2)\), que depende del tiempo \(n\).  
- Es **lineal** y **causal**; es **BIBO estable** porque la suma promediada no puede exceder el m√°ximo de la entrada.  
- **No es TI** (no LTI) porque el multiplicador \(\cos(\pi n/2)\) var√≠a con \(n\): un simple desplazamiento de la entrada no produce un mero desplazamiento de la salida.  
- La respuesta al impulso \(h[n]\) es un **pulso finito** (ventana de longitud \(N\!+\!1\)) ponderado por una **portadora discreta** \(\cos(\pi n/2)\), lo que alterna signos y ceros seg√∫n \(n\).

<summary>Problema 1.1 ‚Äî Propiedades del sistema continuo</summary>

**Enunciado**  
Sea el sistema de tiempo continuo definido por  
$$

$$
Analizar si el sistema presenta **estabilidad**, **invarianza en el tiempo** y **causalidad**.

---

### 1) Estabilidad
Si la entrada est√° acotada, $|x(t)|\le B$ para todo $t$, entonces
$$

\le \int_{t-T}^{t}|x(\tau)|\,d\tau
\le B\int_{t-T}^{t} d\tau
= BT < \infty.
$$
**Conclusi√≥n:** BIBO **estable**.

---

### 2) Invarianza temporal
Para $x'(t)=x(t-t_0)$,
$$
y'(t)=\int_{t-T}^{t}x'(\tau)\,d\tau
=\int_{t-T}^{t}x(\tau-t_0)\,d\tau
=\int_{t-T-t_0}^{t-t_0}x(\sigma)\,d\sigma
= y(t-t_0).
$$
**Conclusi√≥n:** **invariante en el tiempo**.

### 3) Causalidad

$$

$$

**Teor√≠a**
